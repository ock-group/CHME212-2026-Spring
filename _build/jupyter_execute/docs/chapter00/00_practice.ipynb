## üî¨ Step 4: Scientific Libraries

### üéØ NumPy Arrays vs Python Lists

NumPy arrays are much more powerful for scientific computing:

```python
# Convert our temperature list to a NumPy array
temp_array = np.array(temperatures)

print(f"Python list: {temperatures}")
print(f"NumPy array: {temp_array}")
print(f"\nArray statistics:")
print(f"Mean: {np.mean(temp_array):.2f}¬∞C")
print(f"Standard deviation: {np.std(temp_array):.2f}¬∞C")
print(f"Median: {np.median(temp_array):.2f}¬∞C")
```

### üß™ Array Operations (Magic!)

With NumPy, you can do calculations on entire arrays at once:

```python
# Convert Celsius to Fahrenheit for ALL temperatures at once!
temp_fahrenheit = temp_array * 9/5 + 32

print("Temperature conversions:")
for i in range(len(temp_array)):
    print(f"{temp_array[i]:5.1f}¬∞C = {temp_fahrenheit[i]:5.1f}¬∞F")
```

## üìä Step 5: Creating Your First Plot

Let's visualize our temperature data:

```python
# Create time points for our temperature measurements
time_hours = np.arange(0, len(temperatures))  # 0, 1, 2, 3, ...

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(time_hours, temp_array, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (¬∞C)')
plt.title('Reactor Temperature Over Time')
plt.grid(True, alpha=0.3)
plt.show()

print(f"üìà Plot shows temperature rising from {temp_array[0]:.1f}¬∞C to {temp_array[-1]:.1f}¬∞C")
```

## ‚öóÔ∏è Step 6: Real Chemical Engineering Calculation

Let's solve a real ChemE problem: **Ideal Gas Law**

**Problem:** Calculate the pressure of 2.5 moles of nitrogen gas at 350 K in a 10 L container.

**Formula:** PV = nRT  ‚Üí  P = nRT/V

```python
# Given values
n = 2.5      # moles
R = 0.08314  # L‚ãÖbar/(mol‚ãÖK) - ideal gas constant
T = 350      # K
V = 10       # L

# Calculate pressure
P = (n * R * T) / V

print("=== Ideal Gas Law Calculation ===")
print(f"Given:")
print(f"  n = {n} mol")
print(f"  T = {T} K")
print(f"  V = {V} L")
print(f"  R = {R} L‚ãÖbar/(mol‚ãÖK)")
print(f"\nUsing PV = nRT:")
print(f"P = nRT/V = ({n} √ó {R} √ó {T}) / {V}")
print(f"P = {P:.2f} bar")
print(f"P = {P * 14.504:.1f} psi")  # Convert to psi
```

### üéØ Challenge: Multiple Conditions

What if we want to see how pressure changes with temperature? Let's use arrays!

```python
# Create a range of temperatures
T_range = np.linspace(300, 400, 11)  # 300K to 400K, 11 points

# Calculate pressure for each temperature
P_range = (n * R * T_range) / V

# Display results
print("Temperature vs Pressure:")
print("  T (K)    P (bar)")
print("  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
for i in range(len(T_range)):
    print(f"  {T_range[i]:5.0f}    {P_range[i]:6.2f}")
```

### üìà Visualize the Relationship

```python
# Plot pressure vs temperature
plt.figure(figsize=(8, 6))
plt.plot(T_range, P_range, 'r-', linewidth=3, label='P = nRT/V')
plt.scatter([T], [P], color='blue', s=100, zorder=5, label=f'Our calculation\n(T={T}K, P={P:.2f} bar)')
plt.xlabel('Temperature (K)')
plt.ylabel('Pressure (bar)')
plt.title('Ideal Gas Law: Pressure vs Temperature\n(n=2.5 mol, V=10 L)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("üî¨ This shows the linear relationship between temperature and pressure!")
```
