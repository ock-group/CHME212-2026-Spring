## Problem 2: Ideal Gas Law - Rigid Vessel Analysis (15 pts)

A rigid vessel contains 50.0 mol of an ideal gas at an initial temperature of 300 K and a volume of 0.5 m³.

Use the ideal gas law: **PV = nRT**

Where:
- P = pressure (Pa)
- V = volume (m³)
- n = number of moles (mol)
- R = 8.314 J/(mol·K) (universal gas constant)
- T = temperature (K)

### Task 1 (5 pts): Compute the initial pressure

Calculate the initial pressure in these three units:
- Pascals (Pa)
- Kilopascals (kPa)
- Atmospheres (atm) [1 atm = 101,325 Pa]

````python
# Task 1: Initial pressure calculation

V = 0.5 # volume in m^3
n = 50.0 # number of moles
R = 8.314 # J/(mol·K)
T = 300.0 # temperature in K

P_init = (n * R * T) / V
# P_init in Pascals
print(f"Initial Pressure: {P_init} Pa")

# P_init in kPa
P_init_kPa = P_init / 1000
print(f"Initial Pressure: {P_init_kPa} kPa")

# P_init in atm
P_init_atm = P_init / 101325
print(f"Initial Pressure: {P_init_atm} atm")
````

### Task 2 (5 pts): Calculate the final pressure

The temperature is increased linearly to 600 K. Calculate the final pressure in the same three units.

````python
# Task 2: Final pressure calculation

T_final = 600.0 # final temperature in K
P_final = (n * R * T_final) / V
# P_final in Pascals
print(f"Final Pressure: {P_final} Pa")
# P_final in kPa
P_final_kPa = P_final / 1000
print(f"Final Pressure: {P_final_kPa} kPa")
# P_final in atm
P_final_atm = P_final / 101325
print(f"Final Pressure: {P_final_atm} atm")
````

## Problem 3: Flow Rate and Velocity Calculations (10 pts)

Water flows through a circular pipe system with varying diameters:
- Section A: diameter = 0.10 m, velocity = 4.0 m/s
- Section B: diameter = 0.15 m, velocity = ? (to be determined)

For incompressible flow, the continuity equation states: **A₁v₁ = A₂v₂**

Where:
- A = cross-sectional area (m²) = π(D/2)²
- v = velocity (m/s)
- Use π = 3.14

### Task 1. **(5 pts)** Calculate the volumetric flow rate Q in m³/s and L/min

````python
# Your solution here
d1 = 0.1  # diameter in meters
A1 = 3.14 * (d1/2)**2  # cross-sectional area in m^2
v1 = 4.0  # velocity in m/s

Q = A1 * v1  # volumetric flow rate in m^3/s
print(f"Volumetric Flow Rate: {Q} m^3/s")
````

### Task 2. **(5 pts)** Calculate the velocity in Section B using the continuity equation

````python
# Your solution here

d1 = 0.1  # diameter in meters
A1 = 3.14 * (d1/2)**2  # cross-sectional area in m^2
v1 = 4.0  # velocity in m/s

d2 = 0.15 # diameter in meters
A2 = 3.14 * (d2/2)**2  # cross-sectional
v2 = (A1 * v1) / A2  # velocity in m/s
print(f"Velocity in second pipe: {v2} m/s")
````

### Task 2: Calculating the Molecular Weight of a Compound (No Loops or Conditionals) (10 pts)

Write a Python program that calculates the **molecular weight** of a compound containing only carbon (C), hydrogen (H), and oxygen (O), given its **stoichiometric formula** (e.g., `C2H6O`, `CH4`, `C6H12O6`).

Use the following atomic weights:
- Carbon (C) = 12 g/mol  
- Hydrogen (H) = 1 g/mol  
- Oxygen (O) = 16 g/mol  


#### Requirements

1. Assume the formula always has **C first, then H, then O** (Oxygen may be missing).  
2. Use `.find()`, slicing, and `int()` to extract the counts of each atom.  
3. If an element has no number written (e.g., `H` in `CH4`), assume the count is 1.  
4. Print the **molecular weight** as a number in grams per mole.


#### Example Runs

- `C2H4` 
- `C12H26` 
- `C3H6`
- `C2H2`
- `C5H12`

````python
# Your solution here

# Input stoichiometric formula
formula = "C6H12O6" #input("Enter a molecular formula (C, H, O only): ").strip()

# Find element positions
c_pos = formula.find("C")
h_pos = formula.find("H")
o_pos = formula.find("O")

# Extract number of carbon atoms
C = int(formula[c_pos+1:h_pos] or "1")

# Extract number of hydrogen atoms
# If O is present, slice up to O; else slice to the end
H = int(formula[h_pos+1:o_pos]) #if o_pos != -1 else formula[h_pos+1:] or "1")

# Extract number of oxygen atoms
# If O is present, slice from O to end; else 0
O = int(formula[o_pos+1:] or "1") #if o_pos != -1 else 0

# Atomic weights
C_weight = 12
H_weight = 1
O_weight = 16

# Compute molecular weight
molecular_weight = C * C_weight + H * H_weight + O * O_weight

# Output
print("Molecular weight:", molecular_weight, "g/mol")
````
