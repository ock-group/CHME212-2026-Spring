### 1.2 Example 2: For Loop - Calculating Conversion

**Problem:** A batch reactor achieves conversion X at different times. Calculate conversion at t = 0, 5, 10, 15, 20 minutes using:

$$
X(t) = 1 - e^{-kt}
$$

where k = 0.05 minâ»Â¹

**Pseudocode:**
```
SET k = 0.05
SET times = [0, 5, 10, 15, 20]
FOR each time t in times:
    CALCULATE X = 1 - exp(-k*t)
    OUTPUT t and X
END FOR
```

````python


# Example 2: For loop with list iteration

# Rate constant (min^-1)
k = 0.05

# Time points (minutes)
times = [0, 5, 10, 15, 20]

print("Batch Reactor Conversion vs. Time")
print("=" * 40)
print(f"{'Time (min)':<15} {'Conversion':<15}")
print("-" * 40)

# Loop through each time point
for t in times:
    # Calculate conversion using first-order kinetics
    X = 1 - pow(2.71828, -k * t) # or use math.exp(-k * t)
    
    # Print formatted results
    print(f"{t:<15} {X:<15.4f}")
````

### 1.3 Example 3: While Loop - Iterative Calculation !!!skip this part!!!

**Problem:** Calculate the number of iterations needed for a reactor to reach 95% conversion when X increases by 5% each iteration.

**Pseudocode:**
```
SET X = 0
SET target = 0.95
SET iterations = 0
WHILE X < target:
    INCREMENT X by 0.05
    INCREMENT iterations by 1
    OUTPUT current X and iterations
END WHILE
OUTPUT "Target reached in [iterations] iterations"
```

````python
# Example 3: While loop

# Initialize conversion and target
X = 0.0
target_conversion = 0.95
iterations = 0

print("Iterative Conversion Calculation")
print("=" * 40)
print(f"Target conversion: {target_conversion}")
print()

# Continue looping while conversion is below target
while X < target_conversion:
    # Increase conversion by 5% each iteration
    X += 0.05
    
    # Count the iteration
    iterations += 1
    
    # Display progress
    print(f"Iteration {iterations}: X = {X:.2f}")

# Final summary
print("=" * 40)
print(f"Target reached in {iterations} iterations")
````

### 1.4 Example 4: Using `break` - Emergency Shutdown

**Problem:** Monitor reactor pressure readings and trigger an emergency shutdown if pressure exceeds 80 psi. Stop monitoring immediately when the critical threshold is reached.

**Pseudocode:**
```
SET pressure_readings = [45, 52, 58, 63, 71, 78, 85, 92]
SET critical_pressure = 80
FOR each reading in pressure_readings:
    OUTPUT current reading
    IF reading >= critical_pressure:
        OUTPUT "CRITICAL - Emergency shutdown!"
        BREAK
    END IF
END FOR
OUTPUT "Monitoring stopped"
```

````python
# Example 4: Using break to exit a loop early

# Simulated pressure readings from reactor (psi)
pressure_readings = [45, 52, 58, 63, 71, 78, 85, 92]

# Safety threshold (psi)
critical_pressure = 80

print("Reactor Pressure Monitoring System")
print("=" * 45)
print(f"Critical threshold: {critical_pressure} psi\n")

# Monitor each pressure reading
for reading in pressure_readings:
    print(f"  Pressure reading: {reading} psi", end="")
    
    # Check if pressure exceeds critical threshold
    if reading >= critical_pressure:
        print(f"  â† CRITICAL!")
        print(f"\nðŸš¨ EMERGENCY: Pressure {reading} psi exceeds safe limit!")
        print("   Initiating emergency shutdown...")
        break  # Exit the loop immediately
    else:
        print("  âœ“ OK")

print("\nMonitoring stopped.")
````

### Problem 1: For-Else Loop - Finding Valid Operating Condition

**Problem:** Search for the first temperature that gives a reaction rate > 10 mol/(LÂ·s). If not found, use a default.

**Pseudocode:**
```
SET temperatures = [300, 350, 400, 450, 500]
SET target_rate = 10
FOR each T in temperatures:
    CALCULATE rate using Arrhenius equation
    IF rate > target_rate:
        OUTPUT "Found valid T"
        BREAK
    END IF
ELSE:
    OUTPUT "No valid temperature found, using default"
END FOR-ELSE
```

The **else** clause executes only if the loop completes without hitting `break`.

````python
# Problem 1

# Arrhenius parameters
A = 1e8  # Pre-exponential factor (L/(molÂ·s))
Ea = 50000  # Activation energy (J/mol)
R = 8.314  # Gas constant (J/(molÂ·K))

# Test temperatures (K)
temperatures = [300, 350, 400, 450, 500]
target_rate = 10  # Target rate (mol/(LÂ·s))

print("Searching for Valid Operating Temperature")
print("=" * 50)

# Search for temperature that meets criterion
for T in temperatures:
    # Calculate rate using Arrhenius equation: k = A*exp(-Ea/RT)
    rate = A * pow(2.71828, -Ea / (R * T)) # or use math.exp(-Ea / (R * T))
    
    print(f"T = {T} K  â†’  rate = {rate:.2f} mol/(LÂ·s)")
    
    # Check if this temperature meets our criterion
    if rate > target_rate:
        print(f"\nâœ“ Found valid temperature: {T} K")
        operating_temp = T
        break  # Exit the loop early
else:
    # This executes ONLY if loop completed without break
    print("\nâœ— No valid temperature found in range")
    print("Using default temperature: 500 K")
    operating_temp = 500

print(f"\nSelected operating temperature: {operating_temp} K")
````

### Problem 3: For-Else Loop - Finding Optimal Catalyst

You are testing different catalysts to find one that gives a selectivity > 0.85. Test the following catalysts with their selectivities:

| Catalyst | Selectivity |
|----------|-------------|
| Pt/Alâ‚‚Oâ‚ƒ | 0.72        |
| Pd/C     | 0.81        |
| Ni/SiOâ‚‚  | 0.89        |
| Cu/ZnO   | 0.65        |

**Tasks:**
1. Write pseudocode using a for-else structure
2. Store the data in a dictionary: `catalysts = {"Pt/Al2O3": 0.72, "Pd/C": 0.81, ...}`
3. Use a **for loop** to search for a catalyst with selectivity > 0.85. When found, print the catalyst name and **break** the loop. Use the **else clause** to print "No suitable catalyst found" if none meet the criterion. Add descriptive comments throughout your code

**Hint:** You can loop through a dictionary using:
```python
for name, selectivity in catalysts.items():
    # your code here
```

````python
# Problem 2: Finding Optimal Catalyst
# Task 1
# PSEUDOCODE:
# SET catalysts = dictionary with catalyst names and selectivities
# SET target_selectivity = 0.85
# FOR each catalyst_name, selectivity in catalysts:
#     OUTPUT catalyst_name and selectivity
#     IF selectivity > target_selectivity:
#         OUTPUT "Found suitable catalyst!"
#         BREAK
# ELSE (if loop completes without break):
#     OUTPUT "No suitable catalyst found"
# END FOR-ELSE
````

````python
# Task 2

# Store catalyst data in a dictionary
catalysts = {
    "Pt/Al2O3": 0.72,
    "Pd/C": 0.81,
    "Ni/SiO2": 0.89,
    "Cu/ZnO": 0.65
}

# Task 3
# Target selectivity threshold
target_selectivity = 0.85

# Print header
print("Catalyst Screening for Selectivity > 0.85")
print("=" * 50)

# Search for a catalyst that meets the criterion
for name, selectivity in catalysts.items():
    # Print current catalyst being tested
    print(f"Testing {name}: selectivity = {selectivity:.2f}")
    
    # Check if selectivity exceeds target
    if selectivity > target_selectivity:
        print(f"\nâœ“ Found suitable catalyst: {name}")
        print(f"  Selectivity: {selectivity:.2f} > {target_selectivity}")
        selected_catalyst = name
        break  # Exit loop early - we found a good catalyst
else:
    # This block executes ONLY if the loop finished without break
    print("\nâœ— No suitable catalyst found in the dataset")
    selected_catalyst = None

# Final summary
print("=" * 50)
if selected_catalyst:
    print(f"Recommended catalyst: {selected_catalyst}")
else:
    print("Recommendation: Test additional catalysts")
````
