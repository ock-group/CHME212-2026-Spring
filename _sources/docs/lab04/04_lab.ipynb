**Ideal Gas Equation:**

$$

$$

**Van der Waals Equation:**

$$
p = \frac{nRT}{V - nb} - \frac{n^2a}{V^2}
$$

**Dieterici Equation:**

$$
p = \frac{nRT \exp\left(-\frac{na}{RTV}\right)}{V - nb}
$$
