### Solution 1

````python
import numpy as np

# (a) Matrix creation
# 1. 3x3 identity matrix
I3 = np.eye(3)
print("1. Identity (3x3):\n", I3)

# 2. Matrix A and vector b
A = np.array([[3, 1],
              [2, 4]])
b = np.array([9, 14])
print("\n2. A =\n", A)
print("   b =", b)

# (b) Matrix operations
P = np.array([[1, 2],
              [3, 4]])
Q = np.array([[5, 0],
              [1, 3]])

print("\n1. P * Q (element-wise):\n", P * Q)
print("\n2. P @ Q (matrix product):\n", P @ Q)
print("\n3. Q @ P (matrix product):\n", Q @ P)
# 4. P @ Q ≠ Q @ P — matrix multiplication is NOT commutative in general.
#    The order of multiplication matters: swapping A and B changes the result.
print("\n4. P @ Q == Q @ P?", np.array_equal(P @ Q, Q @ P))
print("   Matrix multiplication is NOT commutative in general.")

# (c) Properties of A
det_A = np.linalg.det(A)
print(f"\n1. det(A) = {det_A:.4f}  → unique solution exists: {det_A != 0}")

A_inv = np.linalg.inv(A)
print("\n2. A_inv =\n", A_inv)

check = np.round(A @ A_inv, 10)
print("\n3. A @ A_inv =\n", check, "  (should be identity)")

rank_A = np.linalg.matrix_rank(A)
print(f"\n4. rank(A) = {rank_A}  (full rank: {rank_A == A.shape[0]})")
````

````python
import numpy as np

# (a) 2x2 system: 4x1 - x2 = 7,  2x1 + 3x2 = 1
A_a = np.array([[ 4, -1],
                [ 2,  3]])
b_a = np.array([7, 1])

x_a = np.linalg.solve(A_a, b_a)
print("(a) Solution:", x_a)
print(f"    x1 = {x_a[0]:.4f},  x2 = {x_a[1]:.4f}")
print("    Verification A @ x =", A_a @ x_a, " (expected b =", b_a, ")")

# (b) 3x3 system
A_b = np.array([[ 2,  1, -1],
                [-3, -1,  2],
                [-2,  1,  2]])
b_b = np.array([8, -11, -3])

x_b = np.linalg.solve(A_b, b_b)
print(f"\n(b) Solution: x1={x_b[0]:.4f}, x2={x_b[1]:.4f}, x3={x_b[2]:.4f}")
print("    Verification A @ x =", np.round(A_b @ x_b, 6), " (expected b =", b_b, ")")

# (c) Singular check
M1 = np.array([[2, 4],
               [1, 2]])
M2 = np.array([[1, 2, 3],
               [0, 1, 4],
               [5, 6, 0]])

print(f"\n(c) M1: det = {np.linalg.det(M1):.4f}, rank = {np.linalg.matrix_rank(M1)}")
print(f"    → SINGULAR (row 2 = 0.5 × row 1 — the two equations are not independent)")

print(f"\n    M2: det = {np.linalg.det(M2):.4f}, rank = {np.linalg.matrix_rank(M2)}")
print(f"    → NON-SINGULAR (unique solution exists)")

# Physical meaning of singular matrix:
# A singular matrix means at least two of your equations are linearly dependent —
# one is just a rescaled or combined version of another. This means you have fewer
# independent constraints than unknowns, so the system has either no solution or
# infinitely many solutions. In engineering terms, you are missing a real physical
# constraint (an underdefined process).
````
