### Question 1 — ⏱ ~1 min

Consider the following function definition:

```python
def arrhenius(T, A=1.0, Ea=50000, R=8.314):
    import numpy as np
    return A * np.exp(-Ea / (R * T))
```

What is the return value of `arrhenius(500)` compared to `arrhenius(500, A=2.0)`?

**(A)** They return the same value because `A` does not affect the exponential term.  
**(B)** `arrhenius(500, A=2.0)` returns exactly twice the value of `arrhenius(500)`.  
**(C)** `arrhenius(500, A=2.0)` returns four times the value of `arrhenius(500)`.  
**(D)** A `TypeError` is raised because `A` is a keyword-only argument.

<details>
<summary>Answer</summary>
**(B)** — The function returns $A \cdot e^{-E_a/(RT)}$. Since `A` multiplies the entire expression, doubling `A` from 1.0 to 2.0 doubles the result. The other parameters are unchanged.

</details>

```python
import numpy as np
A = np.array([[1, 1, 1],
              [0.2, 0.5, 0.8],
              [0.1, 0.3, 0.6]])
b = np.array([100, 40, 20])
x = np.linalg.solve(A, b)
```

Before trusting the answer, which check is **most important**?

**(A)** Confirm `np.linalg.det(A) != 0`, indicating a unique solution exists.  
**(B)** Confirm all elements of `x` are positive (physically meaningful flow rates).  
**(C)** Verify by computing `np.allclose(A @ x, b)`.  
**(D)** All of the above are important, in roughly that order.

<details>
<summary>Answer</summary>

**(D)** — A non-zero determinant guarantees a unique solution exists (A). Physical validation checks sign/magnitude (B). Back-substitution `A @ x ≈ b` is the definitive numerical check (C). In practice (C) is the fastest single verification, but all three are good engineering habits.

**(C)** The quarter-circle area formula was applied incorrectly.  
**(D)** The unit square has an area of 1, so only the first quadrant can be sampled.

<details>
<summary>Answer</summary>

</details>

### Question 10 — ⏱ ~1 min

You have 20 reactor yield measurements and want to compute the **sample** standard deviation. Which call is correct, and why?

**(A)** `np.std(yields)` — the default is always correct for experimental data.  
**(B)** `np.std(yields, ddof=1)` — dividing by $N-1$ corrects for the bias introduced by estimating the mean from the same data.  
**(C)** `np.std(yields, ddof=2)` — you must account for two lost degrees of freedom when computing std from a sample.  
**(D)** `np.std(yields) / np.sqrt(20)` — you need to normalize by the sample size.

<details>
<summary>Answer</summary>

**(B)** — NumPy's default (`ddof=0`) divides by $N$, giving the **population** standard deviation. For a sample drawn from a larger population, `ddof=1` divides by $N-1$ (Bessel's correction), which produces an unbiased estimate of the true population variance. One degree of freedom is "used up" when the sample mean is estimated from the same data.

</details>
