````python
## Solution — Problem 6
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

yields = np.array([87.3, 89.1, 85.6, 90.2, 86.8, 88.4, 91.0, 84.9, 87.7, 89.5,
                   86.2, 90.8, 85.1, 88.9, 87.4, 91.3, 86.0, 89.7, 88.1, 87.6,
                   85.8, 90.5, 86.3, 88.6, 89.2, 87.1, 90.0, 85.4, 88.3, 86.7])
spec = 86.5

# Task 1
y_mean = np.mean(yields)
y_std  = np.std(yields, ddof=1)
print(f"Mean : {y_mean:.2f} %")
print(f"Std  : {y_std:.2f} %  (ddof=1)")

dist = stats.norm(loc=y_mean, scale=y_std)
x    = np.linspace(y_mean - 4*y_std, y_mean + 4*y_std, 300)

# Task 2
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

ax = axes[0]
ax.hist(yields, bins=10, density=True, color='steelblue', alpha=0.6, label='Data')
ax.plot(x, dist.pdf(x), 'k-', linewidth=2, label='Normal PDF')
ax.axvline(spec, color='red', linestyle='--', linewidth=1.8, label=f'Spec = {spec}%')
ax.set_xlabel('Yield (%)')
ax.set_ylabel('Density')
ax.set_title('Histogram + Fitted PDF')
ax.legend()

ax = axes[1]
ax.plot(x, dist.cdf(x), 'darkorange', linewidth=2, label='Normal CDF')
ax.axvline(spec, color='red', linestyle='--', linewidth=1.5, label=f'Spec = {spec}%')
p_fail = dist.cdf(spec)
ax.axhline(p_fail, color='tomato', linestyle=':', linewidth=1.5)
ax.plot(spec, p_fail, 'o', color='tomato', markersize=8)
ax.annotate(f'F({spec}) = {p_fail:.3f}',
            xy=(spec, p_fail), xytext=(spec + 0.3, p_fail - 0.08),
            fontsize=10, color='tomato')
ax.set_xlabel('Yield (%)')
ax.set_ylabel('Cumulative probability')
ax.set_title('Fitted CDF')
ax.legend()

plt.suptitle(r'Batch Yield — $\mathcal{N}(\bar{y},\, s^2)$ fit', fontsize=13)
plt.tight_layout()
plt.show()

# Task 3
p_fail_theory = dist.cdf(spec)
p5_yield      = dist.ppf(0.05)
print(f"\nTheoretical failure rate P(Y < {spec}%) : {p_fail_theory*100:.2f} %")
print(f"5th-percentile yield                    : {p5_yield:.2f} %")
````
