## 1. Mean and Variance of the Uniform Distribution

This supplement shows **where the formulas** $\mu = \frac{a+b}{2}$ and $\sigma^2 = \frac{(b-a)^2}{12}$ come from by deriving them from the definition of expected value.

### From Discrete to Continuous: Expected Value

For a **discrete** random variable, the mean (expected value) is:

$$\mu = E[X] = \sum_x x \, P(X = x)$$

This is a weighted average — each possible value $x$ is weighted by its probability $P(X=x)$.

---

For a **continuous** random variable, $P(X = x) = 0$ for any single point. Instead, we work with the probability density function $f(x)$, where:

$$P(x \leq X \leq x + dx) \approx f(x)\,dx$$

The discrete sum becomes an integral:

$$\sum_x x \cdot P(X=x) \quad\longrightarrow\quad \int x \cdot f(x)\,dx$$

So the **expected value of a continuous random variable** is:

$$\boxed{\mu = E[X] = \int_{-\infty}^{\infty} x\, f(x)\,dx}$$

Each term $x \cdot f(x)\,dx$ means:
- $f(x)\,dx$ — the probability mass in the small interval near $x$
- multiplied by $x$ — the contribution of that region to the weighted average

It is a **weighted average of all possible values**, with the PDF as the weight function.

### Mean of the Uniform Distribution

The Uniform distribution on $[a, b]$ has PDF:

$$f(x) = \frac{1}{b - a} \quad \text{for } a \leq x \leq b, \qquad f(x) = 0 \text{ otherwise}$$

So the integral only needs to run from $a$ to $b$:

$$\mu = \int_a^b x\, f(x)\,dx$$

Substituting $f(x) = \frac{1}{b-a}$:

$$\mu = \int_a^b x \cdot \frac{1}{b-a}\,dx = \frac{1}{b-a} \int_a^b x\,dx$$

Evaluate the integral:

$$= \frac{1}{b-a} \cdot \left[\frac{x^2}{2}\right]_a^b = \frac{1}{b-a} \cdot \frac{b^2 - a^2}{2}$$

Factor $b^2 - a^2 = (b-a)(b+a)$:

$$= \frac{1}{b-a} \cdot \frac{(b-a)(b+a)}{2} = \frac{b+a}{2}$$

$$\boxed{\mu = \frac{a + b}{2}}$$

This makes intuitive sense — the mean is the midpoint of the interval.

### Variance of the Uniform Distribution

The variance is defined as:

$$\sigma^2 = E[X^2] - \mu^2$$

### Step 1: Compute $E[X^2]$

$$E[X^2] = \int_a^b x^2\, f(x)\,dx = \frac{1}{b-a} \int_a^b x^2\,dx$$

$$= \frac{1}{b-a} \cdot \left[\frac{x^3}{3}\right]_a^b = \frac{b^3 - a^3}{3(b-a)}$$

Factor $b^3 - a^3 = (b-a)(b^2 + ab + a^2)$:

$$E[X^2] = \frac{b^2 + ab + a^2}{3}$$

### Step 2: Subtract $\mu^2$

$$\mu^2 = \left(\frac{a+b}{2}\right)^2 = \frac{a^2 + 2ab + b^2}{4}$$

$$\sigma^2 = \frac{b^2 + ab + a^2}{3} - \frac{a^2 + 2ab + b^2}{4}$$

Finding a common denominator of 12:

$$= \frac{4(b^2 + ab + a^2) - 3(a^2 + 2ab + b^2)}{12} = \frac{4b^2 + 4ab + 4a^2 - 3a^2 - 6ab - 3b^2}{12}$$

$$= \frac{b^2 - 2ab + a^2}{12} = \frac{(b - a)^2}{12}$$

$$\boxed{\sigma^2 = \frac{(b-a)^2}{12}}$$

### Numerical Verification

Let's confirm both formulas numerically using a large sample from `scipy.stats.uniform`.

```python
import numpy as np
from scipy import stats

a, b = 3.0, 9.0

# Analytical formulas
mu_theory    = (a + b) / 2
var_theory   = (b - a)**2 / 12
sigma_theory = np.sqrt(var_theory)

print("Analytical results")
print(f"  μ        = (a+b)/2          = {mu_theory:.6f}")
print(f"  σ²       = (b-a)²/12        = {var_theory:.6f}")
print(f"  σ        = (b-a)/√12        = {sigma_theory:.6f}")

# scipy.stats confirmation
dist = stats.uniform(loc=a, scale=b - a)
print("\nscipy.stats.uniform")
print(f"  μ        = {dist.mean():.6f}")
print(f"  σ        = {dist.std():.6f}")

# Large-sample numerical check
rng     = np.random.default_rng(seed=0)
samples = rng.uniform(a, b, 1_000_000)
print("\nSample from 1,000,000 draws")
print(f"  Sample μ = {np.mean(samples):.6f}")
print(f"  Sample σ = {np.std(samples, ddof=1):.6f}")
```

## 2. The Normal Distribution Integral — Why There Is No Simple Antiderivative

The Normal (Gaussian) PDF is:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}}\,e^{-(x-\mu)^2/2\sigma^2}$$

Standardizing with $z = (x-\mu)/\sigma$ reduces everything to the **standard normal**:

$$\frac{1}{\sqrt{2\pi}}\,e^{-z^2/2}$$

A natural question: can we integrate this by hand, the way we would integrate $x^2$ or $e^x$?

**The answer is no** — the function $e^{-z^2/2}$ has no antiderivative expressible in elementary functions (polynomials, exponentials, trig, logs, and their compositions). This is a theorem proved by Liouville in the 19th century.

---

### What We Can Do Instead

#### 1. Define a special function for the CDF

Since the integral cannot be written in closed form, we give it a name and define it as a function in its own right.

The **standard normal CDF** is defined as:

$$\Phi(z) = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}}\,e^{-t^2/2}\,dt$$

An equivalent form uses the **error function** $\operatorname{erf}$:

$$\operatorname{erf}(x) = \frac{2}{\sqrt{\pi}}\int_0^x e^{-t^2}\,dt$$

These two are related by:

$$\Phi(z) = \frac{1}{2}\left[1 + \operatorname{erf}\!\left(\frac{z}{\sqrt{2}}\right)\right]$$

Both $\Phi$ and $\operatorname{erf}$ are **not** elementary — they are special functions that are pre-computed numerically and stored in tables (or in `scipy.stats`).

#### 2. Compute definite integrals numerically

Even though the indefinite integral doesn't exist in closed form, **definite integrals** over specific ranges can be evaluated numerically.

For example:

$$P(-1 < Z < 1) = \int_{-1}^{1} \frac{1}{\sqrt{2\pi}}\,e^{-z^2/2}\,dz$$

cannot be written as a simple formula, but it equals:

$$= \Phi(1) - \Phi(-1) = 2\Phi(1) - 1 \approx 0.6827$$

This is where the **68-95-99.7 rule** comes from — it is the numerical result of evaluating $\Phi$.

---

### The One Gaussian Integral You *Can* Do by Hand

There is a famous result:

$$\int_{-\infty}^{\infty} e^{-x^2}\,dx = \sqrt{\pi}$$

This **can** be proved analytically — but the trick is indirect and clever. It does **not** find an antiderivative; it evaluates the *whole* improper integral at once using a double-integral argument.

#### The Trick: Square the Integral and Switch to Polar Coordinates

Let $I = \displaystyle\int_{-\infty}^{\infty} e^{-x^2}\,dx$. We want to find $I$.

**Step 1.** Consider $I^2$:

$$I^2 = \left(\int_{-\infty}^{\infty} e^{-x^2}\,dx\right)\!\left(\int_{-\infty}^{\infty} e^{-y^2}\,dy\right) = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} e^{-(x^2+y^2)}\,dx\,dy$$

We turned a product of two identical 1-D integrals into a single 2-D integral over the entire $xy$-plane.

**Step 2.** Switch to polar coordinates: $x = r\cos\theta$, $y = r\sin\theta$, $x^2 + y^2 = r^2$, and $dx\,dy = r\,dr\,d\theta$.

$$I^2 = \int_0^{2\pi}\int_0^{\infty} e^{-r^2}\, r\,dr\,d\theta$$

**Step 3.** The $\theta$ integral is trivial (integrand has no $\theta$):

$$I^2 = 2\pi \int_0^{\infty} r\,e^{-r^2}\,dr$$

**Step 4.** Substitute $u = r^2$, $du = 2r\,dr$:

$$\int_0^{\infty} r\,e^{-r^2}\,dr = \frac{1}{2}\int_0^{\infty} e^{-u}\,du = \frac{1}{2}\Big[-e^{-u}\Big]_0^{\infty} = \frac{1}{2}$$

**Step 5.** Combine:

$$I^2 = 2\pi \cdot \frac{1}{2} = \pi \qquad \Longrightarrow \qquad I = \sqrt{\pi}$$

$$\boxed{\int_{-\infty}^{\infty} e^{-x^2}\,dx = \sqrt{\pi}}$$

---

### Connecting Back to the Normal PDF

The result above is exactly what ensures the Normal PDF integrates to 1. With the substitution $x = z/\sqrt{2}$:

$$\int_{-\infty}^{\infty} e^{-z^2/2}\,dz = \sqrt{2}\int_{-\infty}^{\infty} e^{-x^2}\,dx = \sqrt{2}\cdot\sqrt{\pi} = \sqrt{2\pi}$$

So the normalizing constant $\frac{1}{\sqrt{2\pi}}$ in the PDF is exactly what makes the total area equal to 1.

| Result | Value |
|--------|-------|
| $\int_{-\infty}^{\infty} e^{-x^2}\,dx$ | $\sqrt{\pi}$ |
| $\int_{-\infty}^{\infty} e^{-z^2/2}\,dz$ | $\sqrt{2\pi}$ |
| $\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-z^2/2}\,dz$ | $1$ ✓ |

Note what the trick **does not give us**: the indefinite integral $\int e^{-x^2}\,dx$, or a formula for $\Phi(z)$ at any specific $z$. For those, numerical methods (or tables) are still required.

### Numerical Verification

```python
import numpy as np
from scipy import stats, special, integrate

# ── 1. Verify the Gaussian integral ──────────────────────────────────────────
I_numerical, _ = integrate.quad(lambda x: np.exp(-x**2), -np.inf, np.inf)
print(f"∫ e^(-x²) dx  from -∞ to ∞ = {I_numerical:.8f}")
print(f"√π                          = {np.sqrt(np.pi):.8f}")

# ── 2. Verify the normalizing constant ───────────────────────────────────────
norm_check, _ = integrate.quad(lambda z: np.exp(-z**2 / 2) / np.sqrt(2 * np.pi),
                                -np.inf, np.inf)
print(f"\n∫ (1/√2π) e^(-z²/2) dz     = {norm_check:.8f}  (should be 1.0)")

# ── 3. The Φ ↔ erf relationship ──────────────────────────────────────────────
print("\nΦ(z) = ½[1 + erf(z/√2)]  —  verification:")
for z in [-2, -1, 0, 1, 2]:
    Phi_stats = stats.norm.cdf(z)
    Phi_erf   = 0.5 * (1 + special.erf(z / np.sqrt(2)))
    print(f"  z = {z:+d}:  Φ = {Phi_stats:.6f},  ½[1+erf(z/√2)] = {Phi_erf:.6f}")

# ── 4. The 68-95-99.7 rule from Φ ────────────────────────────────────────────
print("\n68-95-99.7 rule:")
for k in [1, 2, 3]:
    p = 2 * stats.norm.cdf(k) - 1
    print(f"  P(-{k} < Z < {k}) = 2Φ({k}) - 1 = {p:.4f}  ({100*p:.2f}%)")
```
