## Core Concepts to Watch Out For

### Ch 10 — NumPy

- A NumPy array evaluates operations on **all elements at once** (vectorized) — no `for` loop needed
- `np.array([1, 2, 3]) * 2` → `[2, 4, 6]`; the same on a Python list gives a repeated list, not element-wise math
- **Indexing:** `A[i, j]` is row `i`, column `j` (both zero-based). `A[1, 2]` is the 2nd row, 3rd column.

### Ch 11 — Linear Algebra & Mass Balances

**The core idea:** any system of $n$ linear equations in $n$ unknowns can be written as $A\mathbf{x} = \mathbf{b}$, where each row of $A$ encodes the coefficients of one balance equation. The solution $\mathbf{x} = A^{-1}\mathbf{b}$ exists and is unique as long as $\det(A) \neq 0$ — i.e., the equations are independent.

**Building the system from mass balances:**
- Every balance has the form $(flow\ in) - (flow\ out) = 0$
- The unknowns (flow rates, concentrations) become entries of $\mathbf{x}$
- Their coefficients in each equation become the corresponding row of $A$
- The right-hand side constants become $\mathbf{b}$


**Verification:** substitute $\mathbf{x}$ back into the original equations — the residual $A\mathbf{x} - \mathbf{b}$ must be $\approx \mathbf{0}$. This directly tests whether the solution satisfies every balance simultaneously.

### Ch 12 — Polynomial Fitting & Metrics

| Metric | Formula | Units | Range |
|--------|---------|-------|-------|
| MAE | $\frac{1}{n}\sum\|y_i - \hat y_i\|$ | same as $y$ | $\geq 0$ |
| MSE | $\frac{1}{n}\sum(y_i - \hat y_i)^2$ | $y^2$ | $\geq 0$ |
| RMSE | $\sqrt{\text{MSE}}$ | same as $y$ | $\geq 0$ |
| $R^2$ | $1 - \text{SS}_\text{res}/\text{SS}_\text{tot}$ | dimensionless | $\leq 1$ (best = 1) |

**MAE (Mean Absolute Error)** — the average magnitude of the errors, ignoring direction. Straightforward to interpret: "on average, my prediction is off by this much in the same units as the data." Less sensitive to large outliers because errors are not squared.

**MSE (Mean Squared Error)** — squares each error before averaging, so large mistakes are penalized much more heavily than small ones. Has units of $y^2$, which makes it hard to interpret directly, but it is mathematically convenient (smooth and differentiable everywhere).

**RMSE (Root Mean Squared Error)** — the square root of MSE, restoring the original units of $y$. Comparable to MAE but more sensitive to large errors. A useful rule of thumb: if RMSE $\gg$ MAE, there are a few large outliers dominating the error.

**$R^2$ (Coefficient of Determination)** — measures how much of the variance in the data is explained by the model, relative to simply predicting the mean every time:
$$R^2 = 1 - \frac{\text{SS}_\text{res}}{\text{SS}_\text{tot}} = 1 - \frac{\sum(y_i - \hat y_i)^2}{\sum(y_i - \bar y)^2}$$
- $R^2 = 1$: perfect fit — the model explains all variation
- $R^2 = 0$: the model does no better than predicting the mean $\bar y$ for every point
- $R^2 < 0$: the model is *worse* than the mean (possible for a badly misspecified model)

### Ch 13–14 — Statistics & Monte Carlo

**Why Monte Carlo?** Two complementary reasons:

1. **The nominal calculation is blind to spread.** Evaluating a model at the mean input ($\tau = \mu$) tells you nothing about what happens when inputs vary. The output distribution can have a long tail that causes many failures even when the nominal result passes the spec.

2. **Analytical uncertainty propagation is generally impossible.** For a nonlinear function $y = g(x_1, x_2, \ldots)$ with uncertain inputs, deriving the exact output distribution analytically requires solving integrals that have no closed form. Monte Carlo sidesteps this entirely — just sample the inputs, run the model, and collect the outputs.

### Ch 15 — Numerical Integration & Differentiation

**Trapezoidal rule** approximates the area under a curve by connecting adjacent data points with straight lines and summing the resulting trapezoid areas:
$$\int_a^b f(x)\,dx \approx \sum_{i=0}^{n-1} \frac{x_{i+1}-x_i}{2}\bigl(f_i + f_{i+1}\bigr)$$
- Error is $O(h^2)$: **halving the step size quarters the error**
- The straight-line segments cut corners off a curved function — the gap between the line and the curve is the truncation error; more points reduce this gap faster than linearly

**Finite difference** approximates a derivative from discrete data by replacing the limit with a ratio of finite differences:

| Position | Formula | Accuracy |
|----------|---------|----------|
| Interior | $f'(x_i) \approx \dfrac{f(x_{i+1}) - f(x_{i-1})}{x_{i+1} - x_{i-1}}$ (central) | $O(h^2)$ |
| Left end | $f'(x_0) \approx \dfrac{f(x_1) - f(x_0)}{x_1 - x_0}$ (forward) | $O(h)$ |
| Right end | $f'(x_n) \approx \dfrac{f(x_n) - f(x_{n-1})}{x_n - x_{n-1}}$ (backward) | $O(h)$ |

Central difference is more accurate because it uses symmetric information on both sides of the point; endpoints are less accurate because they can only look in one direction.

### Ch 16 — Root Finding & Newton's Method

**Newton's method** derives from a first-order Taylor expansion — linearize $f$ at the current guess and find where that tangent line crosses zero:
$$f(x) \approx f(x_n) + f'(x_n)(x - x_n) = 0 \quad\Longrightarrow\quad x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$
- **Quadratic convergence:** the error roughly *squares* each step — correct decimal places double with every iteration, far faster than bisection's linear halving
- **When it fails:** if $f'(x_n) \approx 0$ the tangent is nearly flat and the next iterate shoots far away; a poor initial guess can also send the method toward the wrong root
- **Convergence check:** test $|f(x_n)|$ or $|x_{n+1} - x_n|$ — never $|x_n|$ itself, which measures location not progress

**Bisection** is slower but guaranteed — it uses the Intermediate Value Theorem: if $f(a) \cdot f(b) < 0$, a root must exist in $[a, b]$. Each step halves the bracket:
- After $n$ steps: error $\leq (b-a)/2^n$; steps needed for tolerance $\epsilon$: $n \geq \log_2\!\left(\frac{b-a}{\epsilon}\right)$
- Linear convergence ($O(1/2^n)$) — predictable but slow compared to Newton's quadratic convergence

### Ch 17 — Symbolic Math (SymPy)

- Symbols are exact — $\sqrt{2}$ stays $\sqrt{2}$ until you call `.evalf()`
- Key strength: derive analytical expressions (`sp.diff`, `sp.integrate`, `sp.solve`) that a numerical tool cannot

### Ch 18 — ODEs & Euler's Method

**Euler's method** derives from replacing the derivative with a forward finite difference — take one straight-line step in the direction of the current slope:
$$\frac{dy}{dt} \approx \frac{y_{n+1} - y_n}{h} = f(t_n, y_n) \quad\Longrightarrow\quad y_{n+1} = y_n + h\,f(t_n, y_n)$$
- **1st-order accurate** ($O(h)$): halving $h$ halves the global error — much slower improvement than the trapezoidal rule ($O(h^2)$) because Euler uses only the slope at the *start* of the step and ignores how the slope changes across the interval
- The error accumulates step by step: each local approximation is off by $O(h^2)$, and summing over $\sim 1/h$ steps gives a global error of $O(h)$

**FC1.** A student wants to find $\sqrt{2}$ by solving $f(x) = x^2 - 2 = 0$ with `brentq`.  Running the code raises:

```
ValueError: f(a) and f(b) must have different signs
```

```python
from scipy.optimize import brentq

def f(x):
    return x**2 - 2

root = brentq(f, 2, 3)   # <-- what is wrong?
```

(A) `brentq` cannot handle polynomial functions — use `fsolve` instead  
(B) The tolerance is too tight; pass a more lenient `xtol` to fix the error  
(C) The bracket endpoints both give positive values: $f(2) = 2 > 0$ and $f(3) = 7 > 0$ — there is no sign change, so `brentq` cannot locate a root  
(D) The bracket $[2, 3]$ is too small to evaluate the correct answer  

<details>
<summary>▶ Answer</summary>

**(C)** — `brentq` requires $f(a)$ and $f(b)$ to have **opposite signs** (the Intermediate Value Theorem sign-change condition).  With $f(x) = x^2 - 2$, the positive root is $\sqrt{2} \approx 1.41$, which lies entirely outside $[2, 3]$:

```
f(2) = 4 - 2 =  2  (positive)
f(3) = 9 - 2 =  7  (positive)   ← same sign → ValueError
```

Always **plot $f(x)$ first** to visually confirm a sign change before calling `brentq`:

```python
import numpy as np, matplotlib.pyplot as plt
x = np.linspace(0, 3, 200)
plt.plot(x, x**2 - 2); plt.axhline(0, color='k'); plt.grid(); plt.show()

# Valid bracket: f(1) = -1 < 0, f(2) = 2 > 0
root = brentq(f, 1, 2)
```

### Worked Example 1 — Newton's Method by Hand

Find a root of $f(x) = x^3 - 2x - 5$ starting from $x_0 = 2$.

**Step 1 — Newton iteration formula**

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$

**Step 2 — Fill the table**

| $n$ | $x_n$ | $f(x_n)$ | $f'(x_n)$ | $x_{n+1}$ |

### Worked Example 2 — Euler's Method by Hand

Solve $\dfrac{dC_A}{dt} = -k C_A$ with $C_A(0) = 1.0$ mol/L, $k = 0.5$ min$^{-1}$, using $h = 1$ min for $t = 0$ to $3$ min.

**Euler formula:** $C_A^{(n+1)} = C_A^{(n)} + h \cdot f(t_n, C_A^{(n)})$ where $f = -k C_A$

| $n$ | $t_n$ | $C_A^{(n)}$ | $f = -0.5 C_A^{(n)}$ | $C_A^{(n+1)}$ |
|:---:|:-----:|:-----------:|:--------------------:|:-------------:|
| 0 | 0 | 1.000 | $-0.5(1.000) = -0.500$ | $1.000 + 1(-0.500) = 0.500$ |
| 1 | 1 | 0.500 | $-0.5(0.500) = -0.250$ | $0.500 + 1(-0.250) = 0.250$ |
| 2 | 2 | 0.250 | $-0.5(0.250) = -0.125$ | $0.250 + 1(-0.125) = 0.125$ |
| 3 | 3 | — | — | — |

**Exact solution:** $C_A(t) = e^{-0.5t}$

| $t$ | Euler | Exact | Absolute Error |
|:---:|:-----:|:-----:|:--------------:|
| 1 | 0.500 | 0.6065 | 0.1065 |
| 2 | 0.250 | 0.3679 | 0.1179 |
| 3 | 0.125 | 0.2231 | **0.0981** |

**Halving $h$:**  Euler's method is 1st-order accurate, so the global error $\sim O(h)$.  Halving $h$ to 0.5 min would approximately **halve** the error at $t = 3$ min (to $\approx 0.049$).

### Worked Example 3 — Linear Algebra Mass Balance (By Hand via Inverse Matrix)

Three streams mix at a junction. Solve for the flow rates $F_1, F_2, F_3$:

$$F_1 + F_2 + F_3 = 100$$
$$0.30\,F_1 + 0.50\,F_2 + 0.20\,F_3 = 35$$
$$0.10\,F_1 + 0.20\,F_2 + 0.40\,F_3 = 24$$

Write in matrix form $A\mathbf{x} = \mathbf{b}$, so the solution is $\mathbf{x} = A^{-1}\mathbf{b}$.

**Step 1 — Inverse of a 3×3 matrix**

For a $3\times3$ matrix $A$, the inverse is:

$$A^{-1} = \frac{1}{\det(A)}\,C^T$$

where $C^T$ is the **transpose of the cofactor matrix** (also called the adjugate). Each cofactor $C_{ij} = (-1)^{i+j} M_{ij}$, where $M_{ij}$ is the $(i,j)$ **minor** — the determinant of the $2\times2$ submatrix formed by deleting row $i$ and column $j$.

**Step 2 — Compute $\det(A)$ by cofactor expansion along row 1**

$$A = \begin{bmatrix} 1 & 1 & 1 \\ 0.30 & 0.50 & 0.20 \\ 0.10 & 0.20 & 0.40 \end{bmatrix}$$

$$\det(A) = 1\cdot\det\begin{bmatrix}0.50&0.20\\0.20&0.40\end{bmatrix}
           -1\cdot\det\begin{bmatrix}0.30&0.20\\0.10&0.40\end{bmatrix}
           +1\cdot\det\begin{bmatrix}0.30&0.50\\0.10&0.20\end{bmatrix}$$

$$= 1(0.20 - 0.04) - 1(0.12 - 0.02) + 1(0.06 - 0.05)$$

$$= 0.16 - 0.10 + 0.01 = \mathbf{0.07}$$

**Step 3 — Compute the 9 cofactors**

| | $j=1$ | $j=2$ | $j=3$ |
|:---:|:-----:|:-----:|:-----:|
| $i=1$ | $C_{11} = +(0.50{\cdot}0.40 - 0.20{\cdot}0.20) = +0.16$ | $C_{12} = -(0.30{\cdot}0.40 - 0.20{\cdot}0.10) = -0.10$ | $C_{13} = +(0.30{\cdot}0.20 - 0.50{\cdot}0.10) = +0.01$ |
| $i=2$ | $C_{21} = -(1{\cdot}0.40 - 1{\cdot}0.20) = -0.20$ | $C_{22} = +(1{\cdot}0.40 - 1{\cdot}0.10) = +0.30$ | $C_{23} = -(1{\cdot}0.20 - 1{\cdot}0.10) = -0.10$ |
| $i=3$ | $C_{31} = +(1{\cdot}0.20 - 1{\cdot}0.50) = -0.30$ | $C_{32} = -(1{\cdot}0.20 - 1{\cdot}0.30) = +0.10$ | $C_{33} = +(1{\cdot}0.50 - 1{\cdot}0.30) = +0.20$ |

**Step 4 — Form $A^{-1} = \tfrac{1}{0.07}\,C^T$**

$$C^T = \begin{bmatrix} C_{11} & C_{21} & C_{31} \\ C_{12} & C_{22} & C_{32} \\ C_{13} & C_{23} & C_{33} \end{bmatrix} = \begin{bmatrix} 0.16 & -0.20 & -0.30 \\ -0.10 & 0.30 & 0.10 \\ 0.01 & -0.10 & 0.20 \end{bmatrix}$$

$$A^{-1} = \frac{1}{0.07}\begin{bmatrix} 0.16 & -0.20 & -0.30 \\ -0.10 & 0.30 & 0.10 \\ 0.01 & -0.10 & 0.20 \end{bmatrix} = \begin{bmatrix} \tfrac{16}{7} & -\tfrac{20}{7} & -\tfrac{30}{7} \\[4pt] -\tfrac{10}{7} & \tfrac{30}{7} & \tfrac{10}{7} \\[4pt] \tfrac{1}{7} & -\tfrac{10}{7} & \tfrac{20}{7} \end{bmatrix}$$

**Step 5 — Solve $\mathbf{x} = A^{-1}\mathbf{b}$**

$$\begin{bmatrix}F_1\\F_2\\F_3\end{bmatrix} = \frac{1}{7}\begin{bmatrix} 16 & -20 & -30 \\ -10 & 30 & 10 \\ 1 & -10 & 20 \end{bmatrix}\begin{bmatrix}100\\35\\24\end{bmatrix}$$

$$F_1 = \tfrac{1}{7}(16{\cdot}100 - 20{\cdot}35 - 30{\cdot}24) = \tfrac{1}{7}(1600 - 700 - 720) = \tfrac{180}{7} \approx \mathbf{25.71} \text{ kg/h}$$

$$F_2 = \tfrac{1}{7}(-10{\cdot}100 + 30{\cdot}35 + 10{\cdot}24) = \tfrac{1}{7}(-1000 + 1050 + 240) = \tfrac{290}{7} \approx \mathbf{41.43} \text{ kg/h}$$

$$F_3 = \tfrac{1}{7}(1{\cdot}100 - 10{\cdot}35 + 20{\cdot}24) = \tfrac{1}{7}(100 - 350 + 480) = \tfrac{230}{7} \approx \mathbf{32.86} \text{ kg/h}$$

**Step 6 — Verify**

| Equation | Computed | Target |
|----------|----------|--------|
| Overall | $25.71 + 41.43 + 32.86 = 100.00$ | $100$ ✓ |
| Species A | $0.30(25.71) + 0.50(41.43) + 0.20(32.86) = 34.99$ | $35$ ✓ |
| Species B | $0.10(25.71) + 0.20(41.43) + 0.40(32.86) = 24.00$ | $24$ ✓ |
