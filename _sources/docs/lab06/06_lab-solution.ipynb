### Example 1: Matrix Operations

> **Instructor note:** Let's start with the basics. A matrix in NumPy is just a 2D array — you build it with `np.array()` and pass in a list of rows. The most important thing to understand right away is that matrix multiplication is **not** commutative: A×B ≠ B×A in general. This is completely different from regular number multiplication and it trips people up. We use `np.matmul(A, B)` — or equivalently `A @ B` — for matrix multiplication. Watch what happens when I compute AB and BA for these two matrices.

````python
import numpy as np
# Define two matrices
A = np.array([[2, 3], 
              [4, 1]])

B = np.array([[1, 5], 
              [3, 2]])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)

# Matrix multiplication: AB
AB = np.matmul(A, B)
print("\nA × B =")
print(AB)

# Matrix multiplication: BA
BA = np.matmul(B, A)
print("\nB × A =")
print(BA)

print("\nAre AB and BA equal?", np.array_equal(AB, BA))
````

### Example 2: Balancing a Synthesis Reaction using Linear Algebra

> **Instructor note:** Now here's where it gets powerful. You all know how to balance a chemical equation by trial and error — but what if there are 6 or 8 species? Trial and error breaks down. The trick is to recognize that "atoms are conserved" is just a system of linear equations. Each element gives you one equation; each species is one unknown coefficient. Once you fix one coefficient — I'll set a = 1 to anchor the scale — you have as many equations as unknowns, and `np.linalg.solve()` does the rest. Let's start with the simplest possible case: only two elements, three species. This is the Haber process for making ammonia.

Use the linear algebra approach to balance the synthesis of ammonia (Haber process):

$$
N_2 + H_2 \rightarrow NH_3
$$

**Approach:**
1. Assign coefficients: a·N₂ + b·H₂ → c·NH₃
2. Write balance equations for each element (N, H)
3. Set one coefficient (a = 1) to make the system solvable
4. Solve for the remaining coefficients (b, c) using `np.linalg.solve()`
5. Verify the result

````python
import numpy as np

# Example 3: Balancing a Synthesis Reaction using Linear Algebra
# a·N₂ + b·H₂ → c·NH₃
#
# Balance equations for each element (set a = 1):
#   N:  2a = c      →  -c     = -2   →  c = 2
#   H:  2b = 3c     →  2b - 3c = 0
#
# With a=1 fixed, unknowns are b, c — 2 unknowns, 2 equations:
#   Eq1 (N):  0b +  c = 2     (c = 2)
#   Eq2 (H):  2b - 3c = 0     (2b = 3c)

A = np.array([
    [0,  1],   # N balance: c = 2
    [2, -3],   # H balance: 2b - 3c = 0
], dtype=float)

rhs = np.array([2, 0], dtype=float)

# Solve for [b, c]
sol = np.linalg.solve(A, rhs)
b, c = sol
a = 1.0

print("Coefficients:")
print(f"  a = {a:.1f}  (N₂)")
print(f"  b = {b:.1f}  (H₂)")
print(f"  c = {c:.1f}  (NH₃)")

print(f"\nBalanced equation:")
print(f"  {int(a)} N₂ + {int(b)} H₂ → {int(c)} NH₃")

# Verification
print("\nVerification:")
print(f"  N:  LHS = {int(a)*2},  RHS = {int(c)*1}   → {'OK' if a*2 == c else 'FAIL'}")
print(f"  H:  LHS = {int(b)*2},  RHS = {int(c)*3}   → {'OK' if b*2 == c*3 else 'FAIL'}")
````

### Problem 1: Matrix Inverse

> **Instructor note:** The inverse M⁻¹ is the matrix equivalent of dividing by M — multiplying by it "undoes" the matrix. A key fact: not every matrix has an inverse (a matrix is invertible only if its determinant is non-zero). Use `np.linalg.inv()` to compute it, and then verify your answer by checking that both MM⁻¹ and M⁻¹M give the identity matrix. Use `np.allclose()` rather than `==` for the comparison, because floating-point arithmetic introduces tiny rounding errors.

Define matrix `M = np.array([[2, 3], [4, 5]])`. Find its inverse M⁻¹ and check that MM⁻¹ = M⁻¹M = I (identity matrix). Check this in Python.

**Hint:** Use `np.linalg.inv()` to find the inverse matrix.

````python
import numpy as np

# Problem 5: Matrix Inverse
M = np.array([[2, 3],
              [4, 5]])

print("Matrix M:")
print(M)

# Compute the inverse
M_inv = np.linalg.inv(M)
print("\nInverse M⁻¹:")
print(M_inv)

# Verify MM⁻¹ = I
MM_inv = np.matmul(M, M_inv) # M @ M_inv
print("\nM × M⁻¹ =")
print(np.round(MM_inv, 10))

# Verify M⁻¹M = I
M_inv_M = np.matmul(M_inv, M) # M_inv @ M
print("\nM⁻¹ × M =")
print(np.round(M_inv_M, 10))

I = np.eye(2)
print(f"\nMM⁻¹ = I? {np.allclose(MM_inv, I)}")
print(f"M⁻¹M = I? {np.allclose(M_inv_M, I)}")
````

### Problem 2: Solving System of Linear Equations

> **Instructor note:** There are two ways to solve **Ax = b** that we want you to know. `np.linalg.solve()` is what you'd use in practice — it's fast and numerically stable. Do Cramer's rule first by hand (or step by step in code), then confirm with `np.linalg.solve()`. If the two answers don't match, check your determinant calculation first.

Solve the following system of linear equations by hand, and then check the result in Python (using `np.linalg.solve(A, b)`):

$$
\begin{align}
2x + 3y + 4z &= 6 \\
4x + 2y - 3z &= 8 \\
-2x + 5y + 8z &= -4
\end{align}
$$

**Note:** For `np.linalg.solve()`, the coefficient matrix A should contain the coefficients of x, y, z, and the vector b should contain the right-hand side values.

````python
import numpy as np

# Problem 7: Solving System of Linear Equations
# 2x + 3y + 4z = 6
# 4x + 2y - 3z = 8
# -2x + 5y + 8z = -4

A = np.array([[ 2,  3,  4],
              [ 4,  2, -3],
              [-2,  5,  8]], dtype=float)

b_vec = np.array([6, 8, -4], dtype=float)


# Python verification using np.linalg.solve
solution = np.linalg.solve(A, b_vec)
print(f"\nnp.linalg.solve:")
print(f"x = {solution[0]:.4f}")
print(f"y = {solution[1]:.4f}")
print(f"z = {solution[2]:.4f}")

````

### Problem 3: Balancing a Combustion Reaction using Linear Algebra

> **Instructor note:** This is the same workflow as Example 2 — but now you have to set it up yourself. Three elements means a 3×3 matrix. The tricky part is oxygen: it appears as a pure reactant (O₂) **and** inside both CO₂ and H₂O on the product side. When you write the O-balance, make sure every oxygen atom on both sides is accounted for. Set a = 1, move it to the right-hand side, and then build your matrix. If you find that any coefficient comes out negative, double-check which side each species is on.

Use the linear algebra approach to balance the combustion of propane:

$$
C_3H_8 + O_2 \rightarrow CO_2 + H_2O
$$

**Approach:**
1. Assign coefficients: a·C₃H₈ + b·O₂ → c·CO₂ + d·H₂O
2. Write balance equations for each element (C, H, O)
3. Set one coefficient (a = 1) to make the system solvable
4. Solve for the remaining coefficients (b, c, d) using `np.linalg.solve()`
5. Verify the result

````python
import numpy as np

# Example 2: Balancing a Combustion Reaction using Linear Algebra
# a·C₃H₈ + b·O₂ → c·CO₂ + d·H₂O
#
# Balance equations for each element (set a = 1):
#   C:  3a      = c        →  c          = 3   (substituted directly)
#   H:  8a      = 2d       →  2d         = 8   →  d = 4
#   O:  2b      = 2c + d   →  2b - 2c - d = 0
#
# With a=1 fixed, unknowns are b, c, d — 3 unknowns, 3 equations:
#   Eq1 (C):  0b +  c + 0d = 3
#   Eq2 (H):  0b + 0c + 2d = 8
#   Eq3 (O):  2b - 2c -  d = 0

A = np.array([
    [0,  1,  0],   # C balance: c = 3
    [0,  0,  2],   # H balance: 2d = 8
    [2, -2, -1],   # O balance: 2b - 2c - d = 0
], dtype=float)

rhs = np.array([3, 8, 0], dtype=float)

# Solve for [b, c, d]
sol = np.linalg.solve(A, rhs)
b, c, d = sol
a = 1.0

print("Coefficients:")
print(f"  a = {a:.1f}  (C₃H₈)")
print(f"  b = {b:.1f}  (O₂)")
print(f"  c = {c:.1f}  (CO₂)")
print(f"  d = {d:.1f}  (H₂O)")

print(f"\nBalanced equation:")
print(f"  {int(a)} C₃H₈ + {int(b)} O₂ → {int(c)} CO₂ + {int(d)} H₂O")

# Verification
print("\nVerification:")
print(f"  C:  LHS = {int(a)*3},  RHS = {int(c)*1}   → {'OK' if a*3 == c else 'FAIL'}")
print(f"  H:  LHS = {int(a)*8},  RHS = {int(d)*2}   → {'OK' if a*8 == d*2 else 'FAIL'}")
print(f"  O:  LHS = {int(b)*2}, RHS = {int(c)*2 + int(d)*1}   → {'OK' if b*2 == c*2 + d else 'FAIL'}")
````

### Problem 4: Separation Train with Recycle: 5×5 System

> **Instructor note:** This is the capstone problem — a real process flowsheet with five unknown flow rates. The reaction-balancing problems had all the structure handed to you (atoms are conserved → one equation per element). Here you have to read a process description and decide **which** mass balances to write. The key principle is the same: for every unit operation (splitter, mixer), write a total mass balance. For each split ratio given, write a specification equation. Together those give you 5 equations in 5 unknowns, which is a 5×5 linear system. Take your time mapping out the flowsheet before writing a single number down.

A feed stream passes through two splitters in series before entering a mixer that combines it with a known recycle stream. We have **5 unknown flow rates** and can write 5 independent mass balance equations.

```
                    ┌──────────┐           ┌──────────┐           ┌──────────┐
F1 (100 mol/s) ───► │ Split-1  ├──► S1 ──► │ Split-2  ├──► S2 ──► │  Mixer   ├──► F4 (bottoms)
                    └────┬─────┘           └────┬─────┘           └──────────┘
                         │                      │                       ▲
                         ▼                      ▼                       │
                    F2 (product A)         F3 (product B)          F5 = 10 mol/s
                    (40% of F1)            (50% of S1)              (known recycle)
```

**Unknowns:** $F_2, F_3, F_4, S_1, S_2$ (mol/s)

**Known:** $F_1 = 100$ mol/s; $F_5 = 10$ mol/s (recycle); Split-1 sends 40% to $F_2$ (so 60% continues as $S_1$); Split-2 sends 50% of its inlet to $F_3$ (so 50% continues as $S_2$).

**5 equations:**
1. Split-1 specification: $F_2 = 0.40 \cdot F_1 = 40$
2. Split-1 overall balance: $F_1 = F_2 + S_1$ → $F_2 + S_1 = 100$
3. Split-2 specification: $F_3 = 0.50 \cdot S_1$ → $F_3 - 0.5\,S_1 = 0$
4. Split-2 overall balance: $S_1 = F_3 + S_2$ → $F_3 - S_1 + S_2 = 0$
5. Mixer overall balance: $S_2 + F_5 = F_4$ → $-F_4 + S_2 = -10$ (since $F_5 = 10$ is known)

Rewriting as $A\mathbf{x} = \mathbf{b}$ with $\mathbf{x} = [F_2, F_3, F_4, S_1, S_2]^T$:

| Eq. | Equation | $F_2$ | $F_3$ | $F_4$ | $S_1$ | $S_2$ | RHS |
|-----|----------|:---:|:---:|:---:|:---:|:---:|:---:|
| 1 | Split-1 spec: $F_2 = 40$ | 1 | 0 | 0 | 0 | 0 | 40 |
| 2 | Split-1 balance: $F_2 + S_1 = 100$ | 1 | 0 | 0 | 1 | 0 | 100 |
| 3 | Split-2 spec: $F_3 - 0.5\,S_1 = 0$ | 0 | 1 | 0 | -0.5 | 0 | 0 |
| 4 | Split-2 balance: $F_3 - S_1 + S_2 = 0$ | 0 | 1 | 0 | -1 | 1 | 0 |
| 5 | Mixer balance: $-F_4 + S_2 = -10$ | 0 | 0 | -1 | 0 | 1 | -10 |

````python
import numpy as np

# Unknowns: x = [F2, F3, F4, S1, S2]
#
# Eq 1:  F2                         =  40   (split-1 spec: 40% of 100)
# Eq 2:  F2              + S1       = 100   (split-1 balance: F1 = F2 + S1)
# Eq 3:       F3         - 0.5*S1   =   0   (split-2 spec: F3 = 0.5*S1)
# Eq 4:       F3         - S1  + S2 =   0   (split-2 balance: S1 = F3 + S2)
# Eq 5:            -F4         + S2 = -10   (mixer balance: S2 + F5 = F4, F5=10)
#
#        F2   F3    F4    S1    S2    rhs
A = np.array([
    [ 1,   0,    0,    0,    0],   # Eq 1
    [ 1,   0,    0,    1,    0],   # Eq 2
    [ 0,   1,    0,   -0.5,  0],   # Eq 3
    [ 0,   1,    0,   -1,    1],   # Eq 4
    [ 0,   0,   -1,    0,    1],   # Eq 5
], dtype=float)

b = np.array([40, 100, 0, 0, -10], dtype=float)

print("Coefficient matrix A (5×5):")
print(A)
print(f"\ndet(A) = {np.linalg.det(A):.4f}  → unique solution exists")

x = np.linalg.solve(A, b)
F2, F3, F4, S1, S2 = x

print("\n── Solution ──")
print(f"  F2 (product A) = {F2:.2f} mol/s")
print(f"  F3 (product B) = {F3:.2f} mol/s")
print(f"  F4 (bottoms)   = {F4:.2f} mol/s")
print(f"  S1 (internal)  = {S1:.2f} mol/s")
print(f"  S2 (internal)  = {S2:.2f} mol/s")

F1, F5 = 100.0, 10.0
print("\n── Verification ──")
print(f"  Overall: F1 + F5           = {F1 + F5:.2f} mol/s  (total in)")
print(f"           F2 + F3 + F4      = {F2+F3+F4:.2f} mol/s  (total out)  → {'OK' if np.isclose(F1+F5, F2+F3+F4) else 'FAIL'}")
print(f"  Split-1: F2 / F1           = {F2/F1:.2f}       (expected 0.40)  → {'OK' if np.isclose(F2/F1, 0.40) else 'FAIL'}")
print(f"  Split-2: F3 / S1           = {F3/S1:.2f}       (expected 0.50)  → {'OK' if np.isclose(F3/S1, 0.50) else 'FAIL'}")
print(f"  Mixer:   S2 + F5           = {S2+F5:.2f} mol/s  (= F4 = {F4:.2f})  → {'OK' if np.isclose(S2+F5, F4) else 'FAIL'}")
residual = np.max(np.abs(A @ x - b))
print(f"  max|A·x - b|               = {residual:.2e}  ✓")
````
